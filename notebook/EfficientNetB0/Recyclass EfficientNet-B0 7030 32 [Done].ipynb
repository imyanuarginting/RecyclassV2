**Inisialisasi Model Transfer Learning Yaitu MobileNet Ke Dalam Variabel Base Model**

*   weights='imagenet' berarti bahwa kita menggunakan bobot yang sudah dilatih sebelumnya dari dataset 'imagenet'
*   include_top=False berarti bahwa kita tidak menyertakan layer terakhir (top layer)
*   input_shape=(224, 224, 3) berarti bahwa bentuk (shape) yang diharapkan dari gambar yang akan dimasukkan ke dalam model






```python
from keras.applications import EfficientNetB0

base_model = EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3),
)

base_model.trainable = False
```

**Inisialisasi Model Menggunakan Sequential API Agar Memungkinkan Untuk Menumpuk Lapisan-Lapisan Secara Berurutan Untuk Membangun Model Jaringan Saraf**



*   layers.Rescaling(1.0/255.0, input_shape=(224, 224, 3)) berarti bahwa kita menormalisasikan setiap pixel dalam gambar dari 0-255 menjadi 0-1
*   layers.RandomFlip(mode='horizontal') berarti bahwa kita melakukan augmentasi pada tiap gambar dengan model horizontal agar gambar terlihat bervariasi dan berguna juga untuk menaikkan akurasi
*   base_model berarti kita memasukkan model pre-trained MobileNetV2 ke dalam model sekuensial
*   layers.Flatten() berarti bahwa kita melakukan perubahan output dari base_model menjadi vektor satu dimensi. Hal ini diperlukan karena lapisan-lapisan berikutnya dalam model adalah lapisan-lapisan berjenis Dense, yang mengharapkan input dalam bentuk vektor satu dimensi.
*   layers.Dense(DENSE_LAYER, activation='relu') berarti bahwa kita mengatur fungsi aktivasi dan jumlah neuron (unit) yang berasal dari variabel DENSE_LAYER  
*   layers.Dropout(DROPOUT_LAYER) berarti bahwa kita mengatur nilai yang berasal dari variabel DROPOUT_LAYER untuk mencegah overfitting
*   layers.Dense(7, activation='softmax') berarti bahwa kita mengatur fungsi aktivasi dan nilai output layer yaitu 7 yang selaras dengan jumlah kelas yaitu 7 kelas

```python
from keras.models import Model, Sequential
from keras import layers

model = Sequential([
    # layers.Rescaling(1.0/255.0, input_shape=(224, 224, 3)),
    layers.RandomFlip(mode='horizontal', input_shape=(224, 224, 3)),
    base_model,
    layers.Flatten(),
    # layers.Dense(32, activation='relu'),
    # layers.Dropout(0.1),
    layers.Dense(7, activation='softmax'),
])

model.summary()
```




*   optimizer=Adam(learning_rate=0.001) berarti bahwa kita mengatur optimizer dan learning rate untuk mencapai akurasi sebaik mungkin (perlu dilakukan eksperimen dalam mengatur nilai learning_rate untuk mencapai akurasi terbaik)
*   loss='categorical_crossentropy' berarti fungsi kerugian, artinya bahwa kita mengukur seberapa baik dan buruk model memprediksi gambar pada data pelatihan. 'categorical_crossentropy' biasanya digunakan untuk tugas klasifikasi multi-kelas di mana label-labelnya diwakili dalam bentuk one-hot encoding.
*   metrics=['accuracy'] berarti bahwa kita mengukur performa model selama pelatihan dan evaluasi. kita mengukur akurasi (accuracy) sebagai metrik untuk memantau seberapa baik model melakukan klasifikasi.

```python
from keras.optimizers import Adam
from keras import metrics

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy'],
)
```

**Inisialisasi Batch Size dan Image Size**
*   BATCH_SIZE berarti setiap gambar yang ada di latih akan dibagi menjadi per-batch yaitu 16
*   IMAGE_SIZE berarti kita mengatur ukuran gambar menjadi 224*224

```python
BATCH_SIZE = 32
```

**Inisialisasi Variabel Training Dan Validation**
*   label_mode='categorical' berarti kelas gambar ada lebih dari dua kelas. Jika hanya ada dua kelas saja berarti menggunakan 'binary'
*   color_mode='rgb' berarti bahwa kita menentukan mode warna gambar yang digunakan, yaitu RGB (Red, Green, Blue).
*   shuffle=True berarti bahwa dataset akan diacak ulang (shuffle) selama setiap epoch (siklus pelatihan). Ini membantu dalam mencegah model menjadi bias terhadap urutan gambar dalam dataset.
*   seed=42 berarti nilai seed yang digunakan untuk menghasilkan urutan acak jika shuffle diaktifkan (bebas disetting nilai berapa saja)

```python
from keras.utils import image_dataset_from_directory

train = image_dataset_from_directory(
    directory=train_dir,
    label_mode='categorical',
    color_mode='rgb',
    batch_size=BATCH_SIZE,
    image_size=(224, 224),
    shuffle=True,
    seed=42,
)

val = image_dataset_from_directory(
    directory=val_dir,
    label_mode='categorical',
    color_mode='rgb',
    batch_size=BATCH_SIZE,
    image_size=(224, 224),
    shuffle=False,
)
```

**Dilakukan Pelatihan Pada Data Training Dan Data Validation**
*   Epochs=10 berarti ada berapa kali model akan melihat seluruh dataset pelatihan dan validasi selama pelatihan. Dalam hal ini, model akan dilatih selama 10 epoch (10 kali iterasi)
*   verbose=2 berarti kita mengontrol seberapa banyak informasi yang akan ditampilkan selama pelatihan. Nilai 2 berarti akan menampilkan informasi per epoch, seperti progres pelatihan dan metrik evaluasi (contohnya, akurasi dan fungsi kerugian) pada setiap epoch. Nilai 1 akan menampilkan informasi yang lebih terbatas, sedangkan nilai 0 akan mematikan tampilan informasi selama pelatihan

```python
%%time
history = model.fit(
    train,
    validation_data=val,
    epochs=7,
    verbose=2,
)
```

```python
predictions = model.predict(val)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

y_pred = np.argmax(predictions, axis=1)
y_true = np.concatenate([y for x, y in val], axis=0)
y_true = np.argmax(y_true, axis=1)

y_dict = {
    0: 'HDPE',
    1: 'LDPE',
    2: 'Other',
    3: 'PET',
    4: 'PP',
    5: 'PS',
    6: 'PVC',
}
y_pred = np.vectorize(y_dict.get)(y_pred)
y_true =np.vectorize(y_dict.get)(y_true)

cm = confusion_matrix(y_true, y_pred)

print(classification_report(y_true, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=['HDPE', 'LDPE', 'Other', 'PET', 'PP', 'PS', 'PVC'],
    yticklabels=['HDPE', 'LDPE', 'Other', 'PET', 'PP', 'PS', 'PVC'],
)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

**Menampilkan Line Chart Untuk Mengetahui Accuracy dan Loss Pada Data Training dan Validation Dari Tiap Epoch**

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_acc = range(len(acc))
epochs_loss = range(len(loss))

fig, (acc_fig, loss_fig) = plt.subplots(1, 2, figsize=(17, 5))

acc_fig.plot(epochs_acc, acc, '#800000', label='Training accuracy')
acc_fig.plot(epochs_acc, val_acc, '#000080', label='Validation accuracy')
acc_fig.set_title('Training and validation accuracy')
acc_fig.legend(["Training", "Validation"], loc ="lower right")

loss_fig.plot(epochs_loss, loss, '#800000', label='Training loss')
loss_fig.plot(epochs_loss, val_loss, '#000080', label='Validation loss')
loss_fig.set_title('Training and validation loss')
loss_fig.legend(["Training", "Validation"], loc ="upper right")

plt.show()
```
